```python
for column in df.columns:
    sns.boxplot(x = df[column])
    plt.title(f"{column} Boxplot")
    plt.xlabel(column, fontsize = 10)
    plt.ylabel('Oultlier', fontsize = 10)
    plt.show()
```

```python
boxplot_info = {}

for column in df.columns:
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    
    
    min_limit = Q1 - 1.5 * IQR
    max_limit = Q3 + 1.5 * IQR
    
    
    boxplot_info[column] = {
        'Min Limit' :min_limit,
        'Max Limit' :max_limit,
        'IQR' :IQR
    }
    
    
for column, info in boxplot_info.items():
    print(f"column : {column}")
    print(f"Min Limit : {info['Min Limit']}")
    print(f"Max Limit : {info['Max Limit']}")
    print(f"IQR : {info['IQR']}")
    print()
```

```python
continous_data = ['MSSubClass', 'LotFrontage', 'LotArea', 'OverallQual', 'OverallCond',
       'YearBuilt', 'YearRemodAdd', 'MasVnrArea', 'BsmtFinSF1', 'BsmtFinSF2',
       'BsmtUnfSF', 'TotalBsmtSF', '1stFlrSF', '2ndFlrSF', 'LowQualFinSF',
       'GrLivArea', 'BsmtFullBath', 'BsmtHalfBath', 'FullBath', 'HalfBath',
       'BedroomAbvGr', 'KitchenAbvGr', 'TotRmsAbvGrd', 'Fireplaces',
       'GarageYrBlt', 'GarageCars', 'GarageArea', 'WoodDeckSF', 'OpenPorchSF',
       'EnclosedPorch', '3SsnPorch', 'ScreenPorch', 'PoolArea', 'MiscVal',
       'MoSold', 'YrSold']
```

```python
Q1 = df[continous_data].quantile(0.25)
Q3 = df[continous_data].quantile(0.75)
IQR=  Q3 - Q1

min_limit = Q1 - 1.5* IQR
max_limit= Q3 + 1.5 * IQR

df[continous_data] = np.where((df[continous_data] < min_limit) | (df[continous_data] > max_limit),
                             df[continous_data].median(), df[continous_data])
```
